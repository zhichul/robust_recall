```python
from datasets import load_dataset
data = load_dataset('akariasai/PopQA')
```

```python
import numpy as np
import matplotlib.pyplot as plt   # only if you want to plot

# 1) some data — replace with your own
x = np.array(data['test']['s_pop'])
# 2) choose logarithmically-spaced bin edges  
#    → log10(min) … log10(max) → 50 equal steps in log-space
# bin_edges = np.logspace(np.log10(x.min()), np.log10(x.max()), num=50)
bin_edges = [1, 100, 1000, 10000, 100000, 1000000000]

# 3) plain NumPy histogram
hist, edges = np.histogram(x, bins=bin_edges, density=False)

# 4) plotting (optional)
plt.hist(x, bins=bin_edges, density=False)
plt.xscale('log')                     # make the x-axis log so bars look uniform
plt.xlabel('subject popularity (monthly views on wikipedia)')
plt.ylabel('count')
plt.show()
```
